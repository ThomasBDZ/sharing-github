```python
import yfinance as yf

# Extracting stock data for Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1y")  # Gathering one year of historical data

# Resetting the index of the dataframe
tesla_data.reset_index(inplace=True)

# Saving the data (optional, to save as CSV file)
tesla_data.to_csv("tesla_stock_data.csv", index=False)

# Displaying the first five rows of the dataframe
print(tesla_data.head())
```

```python
print(tesla_data.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Function to make a graph
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Closing Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.legend()
    plt.grid(True)
    plt.show()

# Fetching Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1y")  # Gathering one year of historical data

# Resetting the index of the dataframe
tesla_data.reset_index(inplace=True)

# Plotting Tesla stock data with a title
make_graph(tesla_data, title='Tesla Stock Prices')
```
