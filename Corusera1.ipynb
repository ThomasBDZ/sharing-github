Data Science Libraries in Python

- **NumPy**: Fundamental package for numerical computing in Python.
- **Pandas**: Data manipulation and analysis library providing powerful data structures.
- **Matplotlib**: Plotting library for creating visualizations in Python.
- **Seaborn**: Data visualization library based on Matplotlib, providing a higher-level interface.
- **Scikit-learn**: Machine learning library with various algorithms and tools for predictive data analysis.
- **TensorFlow**: Open-source machine learning framework for building and training neural networks.
- **Keras**: High-level neural networks API, often used with TensorFlow for easier prototyping.
- **PyTorch**: Deep learning library known for its flexibility and ease of use.
- **Statsmodels**: Library for estimating statistical models and conducting statistical tests.
- **SciPy**: Scientific computing library with modules for optimization, integration, interpolation, and more.
- **NLTK (Natural Language Toolkit)**: Library for natural language processing tasks.

### Arithmetic Expressions Examples

Arithmetic expressions are fundamental in mathematics and programming. They involve basic mathematical operations such as addition, subtraction, multiplication, and division. Here are some examples of arithmetic expressions:

### Addition
```python
# Adding two numbers

````python
# Multiplication and Addition of Numbers
# Define two numbers
number1 = 10
number2 = 5

# Perform multiplication
result_multiply = number1 * number2
print("Multiplication Result:", result_multiply)  # Output: Multiplication Result: 50

# Perform addition
result_add = number1 + number2
print("Addition Result:", result_add)  # Output: Addition Result: 15
````

````python
# Convert Minutes to Hours
# Define the number of minutes
minutes = 150

# Convert minutes to hours
hours = minutes / 60

# Display the result
print(f"{minutes} minutes is equal to {hours} hours")  # Output: 150 minutes is equal to 2.5 hours
````
